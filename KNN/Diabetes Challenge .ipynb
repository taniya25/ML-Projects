```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
dfx = pd.read_csv('Datasets/Diabetes_XTrain.csv')
dfy = pd.read_csv('Datasets/Diabetes_YTrain.csv')
print(dfx.shape,dfy.shape)
```

```python
X_train = dfx.values
Y_train = dfy.values
```

```python
def dist(x1,x2):#here we will calculate the Euclidean distance between 2 points
    return np.sqrt(sum((x1-x2)**2))

#here we calculate the distance between all k points and store it in a list 
def knn(X,Y,query_point,k=5):
    vals = []
    m = X.shape[0] # first class
    for i in range(m):
        d = dist(query_point,X[i])
        vals.append((d,Y[i])) #appending the distance with the class i.e (distance,class)
        #example : (d1,0),(d2,1),(d3,0)
    vals = sorted(vals)
    #Nearest/First K distances:
    vals = vals[:k]
    vals = np.array(vals)
    #print(vals) #this is 2d array with nearest neighbors with label
    #now we need to check the number of points belonging to each label
    new_vals = np.unique(vals[:,1],return_counts=True)
    print(new_vals)
    #We need to find the index with max frequency 
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return pred
    
```

```python
df = pd.read_csv('Datasets/Diabetes_Xtest.csv')
X_test = df.values
print(X_test.shape)


prediction =["Outcomes"]
for i in range(len(X_test[:,0])):
    pred = int(knn(X_train,Y_train,X_test[i]))
    prediction.append(pred)
```

```python
result = pd.DataFrame(prediction)
result.to_csv('Result.csv',index = False)
```
